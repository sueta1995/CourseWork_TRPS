```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1, 10)
```

```python
y = 2 / x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y) # сам график
plt.xlabel('X') # надпись для оси X
plt.ylabel('Y') # надпись для оси Y
plt.title('Пиздец что это') # название графика
```

```python
fig = plt.figure() # создание объекта фигуры
axes = fig.add_axes([0, 0, 1, 1]) # оси, первые два параметра в списке - начало координат, вторые два - правый верхний угол
axes.plot(x, y) # отрисовать сам график
```

```python
a = np.linspace(0, 10, 1001)
b = np.sin(a)
```

```python
x = np.linspace(0, 10, 1001)
y = np.cos(x)
b1 = np.tan(a)
x1 = np.linspace(1, 10, 1000)
b2 = np.tan(x1)
y1 = 1 / b2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) # теперь axes - массив

for row in axes:
    for col in row:
        col.set_xlabel("X")
        col.set_ylabel("Y")

axes[0][0].plot(a, b)
axes[0][1].plot(x, y)
axes[1][0].plot(a, b1)
axes[1][1].plot(x1, y1)

fig.suptitle("Пиздец какой-то")

plt.tight_layout()
```
